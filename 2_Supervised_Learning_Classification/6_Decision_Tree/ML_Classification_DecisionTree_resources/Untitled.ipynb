```python
import pandas as pd
```

```python
df=pd.read_csv("salaries.csv")
df.head()
```

```python
df.degree.unique()
```

```python
df['degree_number']=df['degree'].map({'bachelors':1,'masters':2})
```

```python
df.head()
```

```python
df.drop('degree',inplace=True,axis='columns')
```

```python
df.head()
```

```python
df_encoded=pd.get_dummies(df,columns=['company','job'],drop_first=True)
```

```python
df
```

```python
from sklearn.tree import DecisionTreeClassifier
X=df_encoded.drop('salary_more_then_100k',axis='columns')
y=df['salary_more_then_100k']
model=DecisionTreeClassifier()
model.fit(X,y)
```

```python
from sklearn.metrics import classification_report

y_pred=model.predict(X)
report=classification_report(y,y_pred)
print(report)
```

```python
from sklearn.tree import export_text
print(export_text(model,feature_names=list(X.columns)))
```
