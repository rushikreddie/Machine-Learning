```python
import pandas as pd
```

```python
from sklearn.datasets import fetch_california_housing
housing = fetch_california_housing()
```

```python
dir(housing)
```

```python
housing.target_names
```

```python
housing.DESCR
```

```python
df=pd.DataFrame(housing.data,columns=housing.feature_names)
```

```python
df.head()
```

```python
df['MedHouseVal']=housing.target
df.head()
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
from sklearn.model_selection import train_test_split
X=df.drop('MedHouseVal',axis="columns")
y=df['MedHouseVal']
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=1)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
import time
```

```python
start=time.time()
model=LinearRegression()
model.fit(X_train,y_train)
end=time.time()

y_pred=model.predict(X_test)
r2s_linear=r2_score(y_test,y_pred)
mse_linear=mean_squared_error(y_test,y_pred)

print('r2s:',r2s_linear,'mse:',mse_linear,'time:',end-start)
```

```python
from sklearn.ensemble import GradientBoostingRegressor

start=time.time()
model=GradientBoostingRegressor()
model.fit(X_train,y_train)
end=time.time()

y_pred=model.predict(X_test)
r2s_gb=r2_score(y_test,y_pred)
mse_gb=mean_squared_error(y_test,y_pred)

print('r2s:',r2s_gb,'mse:',mse_gb,'time:',end-start)
```

```python
from xgboost import XGBRegressor

start=time.time()
model=XGBRegressor()
model.fit(X_train,y_train)
end=time.time()

y_pred=model.predict(X_test)
r2s_xg=r2_score(y_test,y_pred)
mse_xg=mean_squared_error(y_test,y_pred)

print('r2s:',r2s_xg,'mse:',mse_xg,'time:',end-start)
```
